Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations. It's great for creating plots, charts, and graphs to visualize our data.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 4, 9, 16, 25])

# Create a simple plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")

# Show the plot
plt.show()
```

In this example:
- We import `matplotlib.pyplot` as `plt`
- We use `numpy` to create some sample data for the x and y axes.
- `plt.plot(x, y)` creates a line plot using the data.
- `plt.xlabel()`, `plt.ylabel()`, and `plt.title()` add labels to the axes and a title to the plot.
- `plt.show()` displays the plot.


Plot constant graph. A constant function has the same output value for every input value, resulting in a horizontal line.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data for the x-axis
x = np.array([0, 1, 2, 3, 4, 5])

# Create data for a constant y-axis (e.g., y = 3)
y_constant = np.array([3, 3, 3, 3, 3, 3])

# Plot the constant graph
plt.plot(x, y_constant)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Constant Graph (y = 3)")

# Show the plot
plt.show()
```

Draw a linear graph. A linear function has a constant rate of change, resulting in a straight line that is not horizontal or vertical, some inclination with angle.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data for the x-axis
x = np.array([0, 1, 2, 3, 4, 5])

# Create data for a linear y-axis (e.g., y = 2x + 1)
y_linear = 2 * x + 1

# Plot the linear graph
plt.plot(x, y_linear)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Linear Graph (y = 2x + 1)")

# Show the plot
plt.show()
```

## Plot a basic quadratic graph

### Subtask:
Generate code to plot a simple quadratic function.


**Reasoning**:
The subtask is to plot a quadratic graph. I will generate code to import the necessary libraries, create data for a quadratic function, and plot it with labels and a title.



```python
import matplotlib.pyplot as plt
import numpy as np

# Create data for the x-axis using linspace
x = np.linspace(-10, 10, 100)

# Calculate data for a quadratic y-axis (e.g., y = x^2)
y_quadratic = x**2

# Plot the quadratic graph
plt.plot(x, y_quadratic)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Quadratic Graph (y = x^2)")

# Show the plot
plt.show()
```

## Add color to the quadratic graph



**Reasoning**:
The task requires modifying the existing quadratic plot code to change the line color. I will add the `color` argument to the `plt.plot()` function.



```python
import matplotlib.pyplot as plt
import numpy as np

# Create data for the x-axis using linspace
x = np.linspace(-10, 10, 100)

# Calculate data for a quadratic y-axis (e.g., y = x^2)
y_quadratic = x**2

# Plot the quadratic graph with a different color
plt.plot(x, y_quadratic, color='purple')

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Quadratic Graph (y = x^2) - Purple Line")

# Show the plot
plt.show()
```
